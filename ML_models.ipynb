```python
#Importing the required libraries to read,visualize and model the givn dataset files
import pandas as pd
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as mtp
import plotly.graph_objects as go
import plotly.express as px
import missingno as msno 
import warnings
warnings.filterwarnings("ignore")
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn import model_selection
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import RepeatedStratifiedKFold
import nltk
from nltk.tokenize import word_tokenize,RegexpTokenizer
from nltk.stem import WordNetLemmatizer
```

```python
#load features as x and target labels as y from directory.
x=nm.load("x_data.npy",allow_pickle=True)
y=nm.load("y_data.npy",allow_pickle=True)
```

```python
#split our data into testing and training sample.
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x , y, test_size= 0.20, random_state=0) 
```

```python

#Define a function to fit and predict on training and test data sets
def fit_and_predict(model,x_train,y_train,x_test,y_test):
    
    
    # Fitting training sample into model and make predictions using testing sample.
    clf = model
    clf.fit(x_train, y_train)
    predictions=clf.predict(x_test)
    
    
    
    #making report like accuracy score,predictions of ML models.
    print("{}" .format(model))
    print('-'*50)
    print(classification_report(y_test,predictions))
    print('-'*50)
    print('Accuracy of classifier on training set : {}%'.format(round(clf.score(x_train, y_train)*100)))
    print('-'*50)
    print('Accuracy of classifier on  testing set : {}%' .format(round(accuracy_score(y_test,predictions)*100)))
    print('='*70)
    print('='*70)
    return clf
```

```python
#define a function to get predictions of x.
def get_predictions(model, x_test):
    predictions=model.predict(x_test)
    return predictions
```

```python
# Create a list of the machine learning  models to be used for classification and regression.
models=[LogisticRegression(C=1.0),DecisionTreeClassifier(),
        RandomForestClassifier(), SVC(kernel='linear', random_state=0)]
```

```python
# Loop through the list of models and use 'fit_and_predict()' function to trian and make predictions.
prepared_models = []
for model in models:
    prepared_models.append(fit_and_predict(model,x_train, y_train,x_test,y_test))

```

```python
#Loop through the list of prepared_models and make predictions on testing data.
for i in range(4):

    print(prepared_models[i])
    print('*'*50)
    print(get_predictions(prepared_models[i], x_test))
```

```python
#loop through prepared_models to get confusion matrices of models.
for i in range(4):

    y_pred= prepared_models[i].predict(x_test)  
    from sklearn.metrics import confusion_matrix  
    con_mat = confusion_matrix(y_test, y_pred)

    print('classifier :',prepared_models[i])
    print(50*'=')
    print('confusion matrix :-')
    print(con_mat)
    print(70*'*')
```
